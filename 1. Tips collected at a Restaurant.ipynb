## 8. Plot the cumulative Frequency Polygon for "Total Bill" and "Tip" 

```python
value,base = np.histogram(total_bill,bins=20)
cummulative = np.cumsum(value)
plt.plot(base[:-1],cummulative)
plt.grid(axis='x',alpha=0.8)
plt.grid(axis='y',alpha=0.8)
plt.title("Cummulative frequency polygon")
plt.show()
```

```python
value,base = np.histogram(tip,bins=20)
cummulative = np.cumsum(value)
plt.plot(base[:-1],cummulative,color='y')
plt.grid(axis='x',alpha=0.8)
plt.grid(axis='y',alpha=0.8)
plt.title("Cummulative frequency polygon")
plt.show()
```

## 9. Prepare the Frequency Table and Bar Chart for "Size". Summarize and explain your findings 

```python
print("Frequency Table for Size")
size = pd.crosstab(index=tips['size'],columns='count')
print(size)
size.plot.bar(figsize=(15,5),rot =0)
```

first we Created a frequency table for size column in dataset using crosstab() function.

Then using the frequency table we created Bar-Chart.

By creating the table and graph we found out the numbers of the families with different sizes (families containing 

different numbers of peoples) that visited our Resturant!!!!

## 10. Prepare two variable frequency tables for "Size" vs "Tip" and "Size" vs "Total Bill". Summarize and explain your findings.  

```python
# size vs tip
x = pd.crosstab(index = tips['size'], columns= tips['tip'],margins = True)
x.index = ["1 member","2 member","3 member","4 member","5 member","6 member","all"]
x
```

Through this table we get to know that how much tip we recieve from families with different number of peoples.
